```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("bd-dec22-births-deaths-by-region.csv", encoding="cp1252")

print(df.to_string())
```

```python
print(df.columns)
```

```python
import matplotlib.pyplot as plt

# Filter for a specific year
period = 2007

# Filters births for the given year
births_filtered = df[
    (df["Birth_Death"] == "Births") &
    (df["Period"] == period)
].copy()

# Makes each set of data have a uniqe region
births_filtered.set_index("Region", inplace=True)

# Plot
births_filtered["Count"].plot(kind="bar", legend=False)
plt.title(f"Births by Region in {period}")
plt.ylabel("Number of Births")
plt.xlabel("Region")
plt.tight_layout()
plt.show()
```
